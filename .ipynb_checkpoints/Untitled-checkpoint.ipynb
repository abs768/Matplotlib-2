# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Real World Examples

#### Line Graph

```python
g=pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5),dpi=100)

plt.title("Gas prices Over Time (in USD)",fontdict={'fontname':'comic sans MS','fontsize':15})

# plt.plot(g.Year,g.USA,'y-.',label='United States')
# plt.plot(g.Year,g.Canada,'b-.',label='Canada')
# plt.plot(g['Year'],g['South Korea'],'c-.',label='South Korea')

for country in g:
    if country!="Year":
        plt.plot(g.Year,g[country],marker='.',label=country)

plt.xticks(g['Year'][::3])


plt.xlabel('Year',fontdict={'fontname':'comic sans MS','fontsize':15})
plt.ylabel('USD(United States Dollars ($) )',fontdict={'fontname':'comic sans MS','fontsize':15})

plt.legend()

plt.savefig('Gas_Price.png',dpi=150)

plt.show()
```

# Load FIFA data

```python
fi=pd.read_csv('fifa_data.csv')
fi.head(5)
```

### Histograms

```python
list_1=[40,50,60,70,80,90,100]

plt.figure(figsize=(8,5),dpi=100)
plt.hist(fi.Overall,bins=list_1,color='#03fc3d')

plt.xticks(list_1)

plt.title('FIFA 2018',fontdict={'fontname':'Brush Script MT','fontsize':15})

plt.xlabel('Overall Skill Rating of the Players ->',fontdict={'fontname':'Brush Script MT','fontsize':15})
plt.ylabel('Number Of Players in the Skill Range ->',fontdict={'fontname':'Brush Script MT','fontsize':15})

plt.savefig('FIFA_2018_OVERALL.png',dpi=150)

plt.show()
```

### Pie Chart

```python
left=fi.loc[fi['Preferred Foot']=='Left'].count()[14]
right=fi.loc[fi['Preferred Foot']=='Right'].count()[14]

plt.figure(figsize=(6,5),dpi=90)

labels=['Left','Right']
colors=['#03a5fc','#9403fc']

plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f %%')

plt.title('Foot Preference of FIFA players')

plt.show()
```

```python
fi.Weight=[int(x.strip('lbs')) if type(x)== str else x for x in fi.Weight]

plt.figure(figsize=(6,5),dpi=90)

light=fi.loc[(fi.Weight<130)].count()[0]
light_medium=fi.loc[(fi.Weight>=130) & (fi.Weight<160)].count()[0]
medium=fi.loc[(fi.Weight>=160) & (fi.Weight<180)].count()[0]
medium_heavy=fi.loc[(fi.Weight>=180) & (fi.Weight<200)].count()[0]
heavy=fi.loc[fi.Weight>=200].count()[0]

plt.title('Weight Distribution of FIFA players')

weights=[light,light_medium,medium,medium_heavy,heavy]
labels=['Under 130','130-160','160-180','180-200','Over 200']
colors=['#e5eb34','#62eb34','#344ceb','#7d34eb','#eb3434']
explode=(0.4,0,0,0,0.4)

plt.pie(weights,labels=labels,colors=colors,autopct='%.2f %%',pctdistance=0.8,explode=explode)

plt.show()
```

### Box And Whiskers Chart:

```python
plt.style.use('ggplot')

plt.figure(figsize=(5,8))

barcelona=fi.loc[fi.Club=="FC Barcelona"]["Overall"]
arsenal=fi.loc[fi.Club=="Arsenal"]["Overall"]
tottenham=fi.loc[fi.Club=="Tottenham Hotspur"]["Overall"]
labels=['FC Barcelona','Arsenal','Tottenham Hotspur']

boxes=plt.boxplot([barcelona,arsenal,tottenham],labels=labels,patch_artist=True,medianprops={'linewidth':2})

for box in boxes['boxes']:
    #Set Edge Color:
    box.set(color='#3459eb',linewidth=2)
    #Set Fill Color:
    box.set(facecolor='#f6fa07')
    
plt.title('Professional Soccer Team Comparision')
plt.ylabel('FIFA Overall Rating',fontdict={'fontname':'Arial','fontsize':14})

plt.show()
```
